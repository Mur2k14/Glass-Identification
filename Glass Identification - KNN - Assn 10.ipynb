**SCALING THE DATA**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Xval = glasstrain.iloc[:,:9].values
```

```python
Yval = glasstrain[["Type of glass"]].values
```

```python
Xvaltest = glasstest.iloc[:,:9].values
Yvaltest = glasstest[["Type of glass"]].values
```

```python
scaler.fit(Xval)
Xval = scaler.transform(Xval)
```

```python
Xvaltest = scaler.transform(Xvaltest)
```

**APPLYING THE KNN MACHINE LEARNING MODEL TO THE DATASET USING EUCLIDEAN**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=6,p=2,metric="minkowski")
knn.fit(Xval,Yval)
```

```python
Y_pred_test = knn.predict(Xvaltest)
```

```python
print(np.concatenate((Y_pred_test.reshape(len(Y_pred_test),1),Yvaltest.reshape(len(Yvaltest),1)),1))
```

**CHECKING THE ACCURACY SCORE OF THE MODEL**

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(Yvaltest,Y_pred_test)
print(f"The accuracy score of the model is {acc * 100} %")
```
